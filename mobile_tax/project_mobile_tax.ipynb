### Импорт функций 

Импортируем функции, разработанные в предыдущих проектах. Новые функции, для удобства, также будем добавлять в данную ячейку.

```python
# Функция для подсчета количества корзин для гистограммы по правилу Герберта Стерджеса
def bins_num_sturges(column):  
    bins_num = abs(1 + int(math.log2(len(column))))
    if(bins_num) < 200:
        return bins_num
    else:
        return 200
    
# Функция для подсчета количества корзин для гистограммы по правилу Фридмана-Дьякониса     
def bins_num_fr_dk(column): 
    Q1 = column.quantile(0.25) #находим первый квартиль
    Q3 = column.quantile(0.75) #находим третий квартиль
    IQR = Q3 - Q1 #находим межквартильный размах
    bins_num = int(abs(2 * IQR * len(column)**(-1/3)))
    print(bins_num)
    if bins_num < 1:
        return 1
    elif bins_num < 200:
        return bins_num
    else:
        return 200
    
# Функция для отрисовки гистограммы с возможностью выбора размера фигуры и способа подсчета корзин 
def draw_hist(dataframe,column,way_bins,xsize=6,ysize = 4): 
    if (way_bins == 'S'):
        dataframe.hist(
                column,
                bins =  bins_num_sturges(dataframe[column]),
                figsize=(xsize,ysize)
                )          
    elif (way_bins == 'FD'):
        dataframe.hist(
                column,
                bins =  bins_num_fr_dk(dataframe[column]),
                figsize=(xsize,ysize)
              )    
    else:
        way_bins = pd.to_numeric(way_bins) #гистограмма с произвольным числом корзин
        dataframe.hist(
                column,
                bins =  way_bins,
                figsize=(xsize,ysize)
              )
    plt.show() 

# Функция для отрисовки boxplot
def draw_boxplot(dataframe,column):
    dataframe.boxplot(column)
    return plt.show()

# Находим границы "усов"
def find_moustache(column): 
    Q1,Q3 = column.quantile([.25,.75]) #находим первый и третий квартили
    IQR = Q3 - Q1 #находим межквартильный размах
    min_val = Q1 - (IQR * 1.5) 
    max_val = Q3 + (IQR * 1.5)
    
    if(min_val < column.min()):
        if(max_val > column.max()):
            return column.min(),column.max()
        else:
            return column.min(),max_val         
    elif(max_val > column.max()):
            return min_val,column.max()
    else:
        return min_val,max_val
           
# Функция для подсчет количества данных находящимися за пределами "усов" и нахождение процента от общих данных
def num_normalized_data_info(dataframe,column): 
    try:
        min_val,max_val = find_moustache(dataframe[column])
    except:
        return print('Не удалось найти усы в ',column,', проблема с типом данных')
    print('Количество данных из столбца ',column, ', которые считаем выбросами:'
      ,dataframe[((dataframe[column] > max_val) | (dataframe[column] < min_val)) & ~dataframe[column].isna()][column].count()
      ,', процент от общего числа: {:.1%}'
      .format(
          dataframe[(dataframe[column] > max_val) | (dataframe[column] < min_val)][column].count()/dataframe[column].count()
          )
    ) 

# Отбор данных без выбросов 
def num_normalized_data(dataframe,column,with_nan = True): 
    try:
        min_val,max_val = find_moustache(dataframe[column])
    except:
        return print('Не удалось найти усы в ',column,', проблема с типом данных')
    if (with_nan):
        return dataframe[((dataframe[column] <= max_val) & (dataframe[column] >= min_val)) | dataframe[column].isna()] 
    else:
        return dataframe[(dataframe[column] <= max_val) & (dataframe[column] >= min_val)]

# Проверяем, что в финальной выборке нет выбросов
def num_normalized_data_check(dataframe,column,dataframe2): 
    try:
        min_val,max_val = find_moustache(dataframe2[column])# квантили и "усы" берем из изначальных данных
    except:
        return print('Не удалось найти усы в ',column,', проблема с типом данных')
    
    return print('Количество выбросов в столбце ',column, ' в итоговой выборке:',dataframe[((dataframe[column] > max_val) | (dataframe[column] < min_val)) & (~dataframe[column].isna())][column].count());
    
    
# функция для отрисовки диаграммы рассеивания
def draw_scatter(dataframe,xcolumn,ycolumn,xsize = 8, ysize = 6,gridsize = 12):
    dataframe.plot(
    x=xcolumn, 
    y=ycolumn, 
    kind='scatter', 
    figsize=(xsize, ysize), 
    alpha = 0.07, 
    )
    
    return plt.show()

# Поиск дублей с учетом регистра
def reg_duplicates_alert(dataframe,column):
    if(len(dataframe[column].unique()) == len(dataframe[column])):
        return print('Дублей в столбце',column, 'нет.')
    else:
        return print('Дубли,в столбце',column, 'есть.')
```

## Подготовка и анализ данных

### Произведение расчетов

Выделим в отдельный столбец месяц из дат в таблицах **"calls", "internet", "messages"**


```python
calls['month'] = pd.DatetimeIndex(calls['call_date']).month
internet['month'] = pd.DatetimeIndex(internet['session_date']).month
messages['month'] = pd.DatetimeIndex(messages['message_date']).month

```

Выделим сводные таблицы с группировкой по **"user_id"** и **"month"** для упрощения дальнейшего объединения таблиц. Вместе с этим рассчитаем с помощью аргумента **"aggfunc"** :
    - количество сделанных звонков и израсходованных минут разговора по месяцам
    - количество отправленных сообщений по месяцам
    - объем израсходованного интернет-трафика по месяцам  

```python
calls_merge = calls.pivot_table( index = ['user_id','month'], values = 'duration' , aggfunc = ['count','sum'] ).reset_index()
calls_merge.columns = ['user_id','month','calls_num','min_total']
internet_merge = internet.pivot_table( index = ['user_id','month'], values = 'mb_used' , aggfunc = 'sum' ).reset_index()
internet_merge.columns = ['user_id','month','mb_used']
messages_merge = messages.pivot_table( index = ['user_id','month'], values = 'id' , aggfunc = 'count' ).reset_index()
messages_merge.columns = ['user_id','month','message_num']


display(calls_merge.head())
display(internet_merge.head())
display(messages_merge.head())
```

Объединим все таблицы в одну для удобства расчета выручки с каждого пользователя. Таблицы **"calls_merge","internet_merge","messages_merge", "users"** соединим по ключу **"user_id"**. После этого присоединим таблицу **"tariffs"** по ключу **"tariff"**

**"calls_merge","internet_merge","messages_merge"** соединяем с параметром "outher", т.к. нам необходимо, что бы строки попали из обеих соединяемых таблиц. Если этого не сделать, то, например, для "calls_merge - internet.merge" в таблицу не попадут сессии пользователей, которые не совершали звонков. 
**"users"** будем соединять с параметром "left", т.к. нам не интересны пользователи, которые не пользовались ни интернетом, ни сообщениями, ни звонками. 
Интересно, есть ли такие?  Посчитаем разницу между количеством уникальных "user_id" в сводной таблице и в исходной. 

При соединении у нас возникнут NaN, поэтому заполним их нулями.

```python
final_data = calls_merge.merge(internet_merge, how = 'outer')
final_data = final_data.merge(messages_merge, how = 'outer')
final_data = final_data.merge(users, how = 'left')
final_data = final_data.merge(tariffs, how = 'left', left_on = 'tariff', right_on = 'tariff_name')

final_data['calls_num'] = final_data['calls_num'].fillna(0)
final_data['min_total'] = final_data['min_total'].fillna(0)
final_data['mb_used'] = final_data['mb_used'].fillna(0)
final_data['message_num'] = final_data['message_num'].fillna(0)
print(final_data.info())

display(final_data['mb_used'])
final_data['gb_used'] = (final_data['mb_used']/1024).apply(np.ceil)
final_data['gb_used'] = final_data['gb_used'].astype(int)
display(final_data.head(50))
print(final_data.info())
print(final_data['city'].value_counts())



print()
if len(final_data['user_id'].unique()) < len(users['user_id'].unique()):
    print('Среди всех пользователей есть такие, кто не пользуется ни интернетом, ни звонками, ни смс')
else:
    print('Среди всех пользователей нет таких, кто не пользуется ни интернетом, ни звонками, ни смс')
```

Далее посчитаем выручку для каждого пользователя за каждый месяц. 
Напишем функцию, для упрощения расчетов. 

```python
def income_value(column):
    value = column['rub_monthly_fee'] 
    if (column['min_total'] > column['minutes_included']):#проверяем, исрасходовано ли больше включенных в тариф минут
        value += (column['min_total'] - column['minutes_included'])*column['rub_per_minute']
    if (column['message_num'] > column['messages_included']):#проверяем, исрасходовано ли больше включенных в тариф сообщений
        value +=  (column['message_num'] - column['messages_included'])*column['rub_per_message'] 
    if (column['gb_used'] > (column['mb_per_month_included']/1024)):#проверяем, исрасходовано ли больше включенных в тариф гигабайт
        value +=  (column['gb_used'] - (column['mb_per_month_included']/1024))*column['rub_per_gb']
        
    return value

final_data['total_income'] = final_data.apply(income_value,axis = 1)
display(final_data.head())
```

### Проверка гипотез

Для проверки гипотез будем использовать критерий Стьюдента, поскольку при подтверждении гипотез о различии выручки для тарифов , а также для Москвы и остальных регионов нам нужно опеределить, вляется ли это различие статистически значимым или же разница обусловлена случайным разбросом. Критерий значимости возьмем 0,05

#### Гипотеза: Средняя выручка пользователей тарифов "Ультра" и "Смарт" различается

Для проверки этой гипотезы сформируем нулевую и альтернативную гипотезы:

**Нулевая гипотеза:** средняя выручка пользователей тарифов «Ультра» и «Смарт» одинакова

**Альтернативная гипотеза:** средняя выручка пользователей тарифов «Ультра» и «Смарт» различается

Нулевую гипотезу принимаем как равенство выручки для двух тарифов, т.к. нам важно убедиться, что вероятность с которой мы допустим ошибку, если примем альтернативную гипотезу, меньше уровня значимости - 5%.  Если мы отвергнем нулевую гипотезу, то далее мы сможем сделать правильный вывод с вероятностью не менее 95% , относительно того, какой тариф приносит выручки для компании.

Посчитаем среднюю выручку для каждого пользователя. Для расчета будем использовать и медиану и среднее, что бы в случае необходимости сделать более точные выводы.

```python
smart_clear_data = clear_data[clear_data['tariff'] == 'smart'].pivot_table(index = 'user_id',values = 'total_income',aggfunc = 'median')
display(smart_clear_data)
print('Дисперсия выборки для тарифа smart: ',np.var(smart_clear_data,ddof=1))
print('Медиана получаемой выручки с пользователя тарифа smart: ',smart_clear_data.median())
print('Среднее получаемой выручки с пользователя тарифа smart: ',smart_clear_data.mean())

ultra_clear_data = clear_data[clear_data['tariff'] == 'ultra'].pivot_table(index = 'user_id',values = 'total_income',aggfunc = 'median')
display(ultra_clear_data)
print('Дисперсия выборки для тарифа ultra: ',np.var(ultra_clear_data,ddof=1))
print('Медиана получаемой выручки с пользователя тарифа ultra: ',ultra_clear_data.median())
print('Среднее получаемой выручки с пользователя тарифа ultra: ',ultra_clear_data.mean())
```

Сравниваемые выборки независимы - это случайная выборка из генеральной свовокупности, поэтому для проверки гипотезы будем использовать метод ttest_ind.

Т.к. дисперсии выборок явно отличаются более чем на 5%, то при использовании метода ttest_ind будем использовать параметр equal_var=False. 



```python
alpha = 0.05

results = st.ttest_ind(smart_clear_data, ultra_clear_data)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Нулевая гипотеза о равной выручке отвергнута, делаем вывод, что выручки для пользователей тарифов ultra и smart значимо различаются

#### Вывод

Т.к. нулевая гипотеза отвергнута, а среднее и медиана выручки от пользователей тарифа smart больше, то, с вероятностью 99.9% можно утверждать: 

Выручка, получаемая от пользователей тарифов smart и ultra значимо различается: больше выручки приносят пользователи тарифа ultra, в среднем на 900 рублей.    


#### Гипотеза: Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Для проверки  гипотезы сформируем нулевую и альтернативную гипотезы:

Нулевая гипотеза: средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов
Альтернативная гипотеза: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Посчитаем среднюю выручку для каждого пользователя из Москвы и остальных регионов. Для расчета также будем использовать и медиану и среднее.

```python
moscow_clear_data = clear_data[clear_data['city'] == 'Москва'].pivot_table(index = 'user_id',values = 'total_income',aggfunc = 'median')
display(moscow_clear_data)
print('Дисперсия выборки пользователей из Москвы: ',np.var(moscow_clear_data,ddof=1))
print('Медиана получаемой выручки от пользователей из Москвы: ',moscow_clear_data.median())
print('Среднее получаемой выручки от пользователей из Москвы: ',moscow_clear_data.mean())

other_city_clear_data = clear_data[clear_data['city'] != 'Москва'].pivot_table(index = 'user_id',values = 'total_income',aggfunc = 'median')
display(other_city_clear_data)
print('Дисперсия выборки пользователей из других регионов: ',np.var(other_city_clear_data,ddof=1))
print('Медиана получаемой выручки от пользователей из других регионов: ',other_city_clear_data.median())
print('Среднее получаемой выручки от пользователей из других регионов: ',other_city_clear_data.mean())
```

Сравниваемые выборки независимы - это случайная выборка из генеральной свовокупности, поэтому для проверки гипотезы будем использовать метод ttest_ind.

Т.к. дисперсии выборок явно отличаются более чем на 5%, то при использовании метода ttest_ind будем использовать параметр equal_var=False. 



```python
alpha = 0.05

results = st.ttest_ind(moscow_clear_data, other_city_clear_data)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Нулевая гипотеза о равной выручке из Москвы и других регионов не отвергнута. С вероятностью почти в 64% такое или большее различие можно получить случайно, а, значит, для данной выборки нельзя сделать выводы о связи между регионом проживания и выручкой, получаемой от пользователя.   

#### Вывод

Разница выручки среднего и медианы между выборкой для пользователей из Москвы и других городов статистически незначима, значит, разница может быть обусловлена случайным отклонением - связь между населенным пунктом и выручкой не подтверждена.
