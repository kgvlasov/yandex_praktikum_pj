Теперь обработаем пропуски в столбцах **"living_area", "is_apartment"** и **"balcony"**. Для каждого столбца проверим, как изменилась гистограмма и информация, полученная методом describe() до обработки пропусков и после. 

Первым обработаем столбец **"living_area"** - заполним данные медианой жилой площади для объектов с соответсвующим количеством комнат.   
Для упрощения задачи в дальнейшем напишем функцию для отрисовки гистограммы и функцию для определения количества корзин разными способами: Стерджеса, Фридмана-Дьякониса и выбор произвольного числа корзин.

```python
#Функция для подсчета количества корзин для гистограммы по правилу Герберта Стерджеса
def bins_num_sturges(column):  
    bins_num = abs(1 + int(math.log2(len(column))))
    if(bins_num) < 200:
        return bins_num
    else:
        return 200
    
#Функция для подсчета количества корзин для гистограммы по правилу Фридмана-Дьякониса     
def bins_num_fr_dk(column): 
    Q1 = column.quantile(0.25) #находим первый квартиль
    Q3 = column.quantile(0.75) #находим третий квартиль
    IQR = Q3 - Q1 #находим межквартильный размах
    bins_num = int(abs(2 * IQR * len(column)**(-1/3)))
    print(bins_num)
    if bins_num < 1:
        return 1
    elif bins_num < 200:
        return bins_num
    else:
        return 200
    
# Функция для отрисовки гистограммы с возможностью выбора размера фигуры и способа подсчета корзин 
def draw_hist(dataframe,column,way_bins,xsize=6,ysize = 4): 
    if (way_bins == 'S'):
        dataframe.hist(
                column,
                bins =  bins_num_sturges(dataframe[column]),
                figsize=(xsize,ysize)
                )          
    elif (way_bins == 'FD'):
        dataframe.hist(
                column,
                bins =  bins_num_fr_dk(dataframe[column]),
                figsize=(xsize,ysize)
              )    
    else:
        way_bins = pd.to_numeric(way_bins) #гистограмма с произвольным числом корзин
        dataframe.hist(
                column,
                bins =  way_bins,
                figsize=(xsize,ysize)
              )
    plt.show()    
    

print('Информация о данных до обработки\r\n',estate_data['living_area'].describe(),'\r\n')
draw_hist(estate_data,'living_area','FD')# отрисовка гистограммы с числом корзин, посчитанным функцией bins_num_fr_dk
draw_hist(estate_data,'living_area','S') # отрисовка гистограммы с числом корзин, посчитанным функцией bins_num_sturges
draw_hist(estate_data,'living_area','30') #отрисовка гистограммы с числом корзин, равным 40

estate_data['living_area'] = estate_data.groupby('rooms')['living_area'].apply(lambda x: x.fillna(x.median()))

print('Информация о данных после обработки\r\n',estate_data['living_area'].describe(),'\r\n')
draw_hist(estate_data,'living_area','30')
```

При анализе гистограмм до и после обработки видно, что после заполнения пропусков распределение изменилось не значительно. Показатели, посчитанные методом describe() также практически не изменились. Значит, заполнение пропусков выполнено верно. 

Распределение не похоже на распределение Гаусса (нормальное распределение) - имеется тяжелый хвост справа. Также распределение не может быть распределеним Пуассона, т.к. оно характеризует число событий в фиксированную единицу времени. Вероятно, это связано с тем, что жилая площадь точно не является случайной величиной. 

Интересно, что несмотря на внешние источники, утверждающие, что способ Фридмана-Дьякониса для подсчета корзин самый надежный, гистограмма, с числом корзин, подсчитанных им, совершенно не информативна. Возможно дело в очень маленьком межквартильном расстоянии. Проверим это на других столбцах. Самой информативной стала гистограмма с 30-ю корзинами (получено эмпирически).

Данных, в которых могут быть выбросы очень много, поэтому поступим следующим образом: сформируем датасет из нормализованных данных, посмотрим на его объем и примем решение - либо использовать только его в дальнейших исследованиях либо проведем исследование как на общих данных, так и на нормализованных и опишем разницу между ними. 


Напишем:

1) фукнцию, отрисовывающую boxplot  для всех данных

2) функцию, возвращающую минимальную и максимальную границы выборки (границы "усов" на диаграмме размаха) - с помощью нее будем "резать" необходимые нам данные.



```python
def draw_boxplot(dataframe,column):
    dataframe.boxplot(column)
    return plt.show()

def find_moustache(column): #Ищем "усы" =)
    Q1,Q3 = column.quantile([.25,.75]) #находим первый и третий квартили
    IQR = Q3 - Q1 #находим межквартильный размах
    min_val = Q1 - (IQR * 1.5) 
    max_val = Q3 + (IQR * 1.5)
   
    if(min_val < column.min()):
        if(max_val > column.max()):
            return column.min(),column.max()
        else:
            return column.min(),max_val         
    elif(max_val > column.max()):
            return min_val,column.max()
    else:
        return min_val,max_val
           
        
```

Построим диаграммы размаха по интересующим нас признакам. 

```python
columns_data = ['last_price','rooms','ceiling_height','living_area','total_area','kitchen_area'
    ,'parks_around3000','ponds_around3000','days_exposition','airports_nearest','cityCenters_nearest']



for each in columns_data:
    draw_boxplot(estate_data,each)

#Функция для подсчет количества данных находящимися за пределами "усов" и нахождение процента от общих данных
def num_normalized_data_info(dataframe,column): 
    min_val,max_val = find_moustache(estate_data[column])# квантили и "усы" берем из изначальных данных
    
    print('Количество данных из столбца ',column, ', которые считаем выбросами:'
      ,dataframe[(dataframe[column] > max_val) | (dataframe[column] < min_val)][column].count()
      ,', процент от общего числа: {:.1%}'
      .format(
          dataframe[(dataframe[column] > max_val) | (dataframe[column] < min_val)][column].count()/dataframe[column].count()
          )
    )

    
for each in columns_data:
    num_normalized_data_info(estate_data,each)               
```

Выбросов, на первый взгляд, не много - не более 10% по каждому признаку. Однако, если отсечь сразу все выбросы, данных может не остаться.

Попробуем отсечь данные и проверим размер выборки - для это напишем функцию.
Далее проверим, остануться ли в финальной выборке изначальные выбросы


```python
columns_data = ['last_price','rooms','ceiling_height','living_area','total_area','kitchen_area'
    ,'days_exposition',]

def num_normalized_data(dataframe,column): 
    min_val,max_val = find_moustache(dataframe[column])
    return dataframe[(dataframe[column] < max_val) & (dataframe[column] > min_val)]



final_data = estate_data

for each in columns_data:
    print('Количество строк в нормализованной выборке',final_data['last_price'].count(),'перед шагом ',each)
    final_data = num_normalized_data(final_data,each)
      
final_data = final_data.reset_index(drop=True)

print()
#Проверяем, что в финальной выборке нет выбросов
def num_normalized_data_check(dataframe,column): 
    min_val,max_val = find_moustache(estate_data[column])
    
    print('Количество выбросов в столбце ',column, ' в итоговой выборке:'
      ,dataframe[(dataframe[column] > max_val) | (dataframe[column] < min_val)][column].count()
      )

final_data = final_data[(final_data['airports_nearest'] > 0) & (final_data['cityCenters_nearest'] > 0)]    
    
for each in columns_data:
    num_normalized_data_check(final_data,each) 

print()
print ('Количество строк в нормализованной выборке',final_data['last_price'].count(),', процент оставшихся данных: {:.1%}'.format(final_data['last_price'].count()/estate_data['last_price'].count()))    


```
